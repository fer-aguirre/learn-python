### Datatypes

```python
import pandas as pd
import numpy as np
```

```python
# !wget https://www.kaggle.com/volpatto/coffee-quality-database-from-cqi?select=merged_data_cleaned.csv
df = pd.read_csv('../../data/coffee_data.csv')
```

### `copy()`: Make a copy of this objectâ€™s indices and data

```python
df_coffee = df.copy(deep=True)
```

### `info()`: Print a concise summary of a DataFrame

```python
df_coffee.info()
```

### `size`: Return an int representing the number of elements in this object

```python
df_coffee.size
```

### `shape`: Return a tuple representing the dimensionality of the DataFrame

```python
df_coffee.shape
```

### `axes`: Return a list representing the axes of the DataFrame

```python
df_coffee.axes
```

### `values`: Return a Numpy representation of the DataFrame

```python
df_coffee.values
```

### `head()`: Returns first *n* rows

```python
df_coffee.head(3)
```

### `tail()`: Returns last *n* rows

```python
df_coffee.tail(3)
```

### `sample()`: Return a random sample of items from an axis of object

```python
df_coffee.sample(3)
```

### `describe()`: Generate descriptive statistics

```python
df_coffee.describe()
```

```python
pd.options.display.float_format = '{:,.1f}'.format
df_coffee.describe(include='all')
```

### `dtypes`: Return the dtypes in the DataFrame

```python
df_coffee.dtypes
```

### `convert_dtypes()`: Convert columns to best possible dtypes using dtypes supporting

```python
df_coffee.convert_dtypes().dtypes
```

### `nunique()`: Count number of distinct elements in specified axis

```python
df_coffee.nunique()
```

### `astype()`: Cast a pandas object to a specified dtype 

```python
df_coffee[['Species', 'Country.of.Origin']] = df_coffee[['Species', 'Country.of.Origin']].astype('category')
```

```python
df_coffee.dtypes
```

### `unique()`: Uniques are returned in order of appearance

```python
df_coffee['Species'].unique()
```

### `value_counts()`: Return a Series containing counts of unique values

```python
df_coffee['Species'].value_counts()
```

### `get_dummies()`: Convert categorical variable into dummy/indicator variables

```python
pd.get_dummies(df_coffee['Species'])
```

### Create columns based on `dummy values` (variable that takes values of 0 for absence and 1 for presence of something)

```python
df_coffee[['Arabica', 'Robusta']] = pd.get_dummies(df_coffee['Species'])
df_coffee.sample(3)
```

### Create a new column with specified value

```python
df_coffee['ones'] = 1
df_coffee.sample(3)
```

### Generate random dates

```python
min_date = pd.to_datetime('2000-01-01')
max_date = pd.to_datetime('2020-12-31')

d = (max_date - min_date).days + 1

df_coffee['Dates'] = min_date + pd.to_timedelta(np.random.randint(d,size=1339), unit='d')
df_coffee.sample(3)
```

### `to_datetime()`: Convert argument to datetime

```python
pd.to_datetime(df_coffee['Dates'], errors='coerce', format='%Y/%m/%d')
df_coffee.sample(3)
```

```python
df_coffee.convert_dtypes().dtypes
```

### `rename()`: Alter axes labels

```python
df_coffee.rename(columns={'Country.of.Origin': 'Country'}, inplace=True)
df_coffee.columns
```

### Convert columns to a list

```python
list(df_coffee)
```

### `drop()`: Drop specified labels from rows or columns

```python
df_coffee.drop(['ones'], axis=1, inplace=True)
df_coffee.head(3)
```

```python
df_coffee.drop([0, 1], inplace=True)
df_coffee.head(3)
```

```python
df_coffee.drop(columns=['Lot.Number', 'Mill'], index=[2, 3], inplace=True)
df_coffee.head(3)
```

### `duplicated()`: Return boolean Series denoting duplicate rows

```python
df_coffee.duplicated(subset=['Farm.Name'])
```

```python
df_coffee.duplicated(keep='first', subset=['Species'])
```

```python
df_coffee.duplicated(keep='last', subset=['Species'])
```
