### Language Stadistics

```python
import matplotlib.pyplot as plt
import numpy as np
from nltk.book import *

import nltk

nltk.download('book')
```

### Create a dictionary where words are the keys and the values are their frequencies

```python
dic = {}
for word in set(text1):
    dic[word] = text1.count(word)
dic
```

### `FreqDist()`: A method to count words on a text

```python
fdist = FreqDist(text1)
```

### `most_common()`: Return the most common frequency words

```python
fdist.most_common(10)
```

### `keys()`: Return a list of all the words in the text

```python
text_words = [word for word in fdist.keys() if len(word) > 15]
text_words
```

### `values()`: Return a list of all the counts of words in the text

```python
test = [value for value in fdist.values() if value > 5000]
test
```

### `plot()`: Plot samples from the frequency distribution displaying the most frequent sample first

```python
fdist.plot(10)
```

### `max()`: The greatest number of outcomes in this frequency distribution

```python
fdist.max()
```

### `['word']`: Return the frequency distribution of the specified word

```python
fdist['monster']
```

### `N()`: Return the total number of sample outcomes that have been recorded by this FreqDist, in other words, the legth of the text

```python
fdist.N()
```

### `B()`: Return the total number of sample values (bins) that have counts greater than zero, in other words, the length of vocabulary

```python
fdist.B()
```

### `tabulate()`: Tabulate the given samples from the frequency distribution (cumulative), displaying the most frequent sample first

```python
fdist.tabulate(10)
```

### `freq()`: Frequency of a sample is defined as the count of that sample divided by the total number of sample outcomes that have been recorded by the FreqDist

```python
fdist.freq('the')
```

### Filter words by length

```python
long_words = [word for word in text1 if len(word) > 5]
long_words[:10]
```

```python
vocabulary_filtered = sorted(long_words)
vocabulary_filtered[:10]
```

```python
fdist_filtered = FreqDist(vocabulary_filtered)
```

```python
fdist_filtered.max()
```

```python
interesting_words = [(word, fdist[word]) for word in set(text1) if len(word) > 5 and fdist[word] > 10]
interesting_words[:10]
```

### Convert list of tuples to a numpy object (array)

```python
dtype = [('word', 'S10'), ('frequency', int)]
interesting_words = np.array(interesting_words, dtype=dtype)
```

```python
interesting_words
```

```python
interesting_words = np.sort(interesting_words, order='frequency')
interesting_words
```

```python
top_words = 10
x = np.arange(len(interesting_words[-top_words:]))
y = [freq[1] for freq in interesting_words[-top_words:]]
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xticks(x, [str(freq[0]) for freq in interesting_words[-top_words:]], rotation='vertical')
plt.grid(True)
plt.show()
```
