# List comprehension

List of the first 100 natural numbers squared

```python
# For loop
squares = []
for i in range(1, 101):
    squares.append(i**2)

# Element for element in iterable
# Each element to append to the new list + from certain loop of elements
square_numbers = [i**2 for i in range(1, 101)]
square_numbers
```

List of the first 100 natural numbers squared if they aren't divisible by 3

```python
# For loop
squares = []
for i in range(1, 101):
    if i%3 != 0:
        squares.append(i**2)

# Element for element in iterable if condition
# Each element to append to the new list + from certain loop of elements + condition to filter elements from loop
square_numbers = [i**2 for i in range(1, 101) if i % 3 != 0]
square_numbers
```

List of the multiples of 4, 6 and 9 up to 5 digits

```python
# Multiples from 4, 6 and 9 
multiples = [i for i in range(1, 99999) if i % 4 == 0 and i % 6 == 0 and i % 9 == 0]
multiples
```
