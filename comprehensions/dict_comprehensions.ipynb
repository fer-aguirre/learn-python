# Dict comprehension

Dictionary of the first 100 natural numbers as keys and their cubes as values

```python
# For loop
my_dict = {}
for i in range(1, 101):
    my_dict[i] = i**3

# Key:value for value in iterable 
cube_numbers = {i: i**3 for i in range(1, 101)}
cube_numbers
```

Dictionary of the first 100 natural numbers as keys and their cubes as values if they aren't divisible by 3

```python
# For loop
my_dict = {}
for i in range(1, 101):
    if i % 3 != 0:
        my_dict[i] = i**3

# Key:value for value in iterable if condition
cube_numbers = {i: i**3 for i in range(1, 101) if i % 3 != 0}
cube_numbers
```

Dictionary of the first 1000 natural numbers as keys and their square roots as values

```python
square_roots = {i: i**0.5 for i in range(1, 1001)}
square_roots
```
